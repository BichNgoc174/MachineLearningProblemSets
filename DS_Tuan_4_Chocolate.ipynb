```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= 'E:/Data_Science/Data_learn/chocolate.csv'
df = pd.read_csv(data)
```

```python
for i in df.columns.values:
    if i != 'Rating':
        df[i] = df[i].apply(lambda x: str(x).replace(u'\xa0', ''))

        
print("So luong: " + str(df.shape[0]))
print("Xep hang trung binh: " + str(df['Rating'].mean()))
df.head()
```

```python
sns.countplot(df['Rating']).set_title('Phan phoi tren xep hang Chocolate')
plt.show()
```

```python
df['Bean\nType'].fillna("Uknown", inplace = True)
df['Bean\nType'].replace("nan", "Uknown", inplace = True)

print("Tong so loai dau duy nhat: " + str(len(df['Bean\nType'].unique())))

f = {'Rating':['size','mean','std']}
df.groupby('Bean\nType').agg(f)
```

```python
print("Tong so loai dau duy nhat: " + str(len(df['Company\nLocation'].unique())))

f = {'Rating':['size','mean','std']}
df.groupby('Company\nLocation').agg(f)
```
